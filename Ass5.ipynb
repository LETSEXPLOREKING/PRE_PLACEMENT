```python
<aside>
ðŸ’¡ **Question 1**

Convert 1D Array Into 2D Array

You are given a **0-indexed** 1-dimensional (1D) integer array original, and two integers, m and n. You are tasked with creating a 2-dimensional (2D) array with Â m rows and n columns using **all** the elements from original.

The elements from indices 0 to n - 1 (**inclusive**) of original should form the first row of the constructed 2D array, the elements from indices n to 2 * n - 1 (**inclusive**) should form the second row of the constructed 2D array, and so on.

Return *an* m x n *2D array constructed according to the above procedure, or an empty 2D array if it is impossible*.

</aside>
```

```python
def convert_to_2d_array(original, m, n):
    if m * n != len(original):
        # If the number of elements in the original array does not match the required size of the 2D array, it is impossible.
        return []

    # Initialize the 2D array with zeros
    result = [[0 for _ in range(n)] for _ in range(m)]

    for i in range(m):
        for j in range(n):
            # Calculate the index in the original array corresponding to the element in the 2D array.
            index = i * n + j
            result[i][j] = original[index]

    return result

# Test the function
original = [1, 2, 3, 4, 5, 6]
m = 2
n = 3
print(convert_to_2d_array(original, m, n))
```

```python
<aside>
ðŸ’¡ **Question 2**

You have n coins and you want to build a staircase with these coins. The staircase consists of k rows where the ith row has exactly i coins. The last row of the staircase **may be** incomplete.

Given the integer n, return *the number of **complete rows** of the staircase you will build*.

</aside>
```

```python
def complete_rows(n):
    # Initialize k as 0
    k = 0

    # Loop to find the largest value of k such that k(k+1)/2 <= n
    while (k + 1) * k // 2 <= n:
        k += 1

    # Since the loop will overshoot by 1, subtract 1 to get the correct value of k
    return k - 1

# Test the function
n = 14
print(complete_rows(n))  # Output: 4
```

```python
<aside>
ðŸ’¡ **Question 3**

Given an integer array nums sorted in **non-decreasing** order, return *an array of **the squares of each number** sorted in non-decreasing order*.

**Example 1:**

**Input:** nums = [-4,-1,0,3,10]

**Output:** [0,1,9,16,100]

**Explanation:** After squaring, the array becomes [16,1,0,9,100].

After sorting, it becomes [0,1,9,16,100].

</aside>
```

```python
def sorted_squares(nums):
    # Square each element in the array
    squared_nums = [num ** 2 for num in nums]

    # Sort the squared numbers in non-decreasing order
    squared_nums.sort()

    return squared_nums

# Test the function
nums = [-4, -1, 0, 3, 10]
print(sorted_squares(nums))  # Output: [0, 1, 9, 16, 100]
```

```python
<aside>
ðŸ’¡ **Question 4**

Given two **0-indexed** integer arrays nums1 and nums2, return *a list* answer *of size* 2 *where:*

- answer[0] *is a list of all **distinct** integers in* nums1 *which are **not** present in* nums2*.*
- answer[1] *is a list of all **distinct** integers in* nums2 *which are **not** present in* nums1.

**Note** that the integers in the lists may be returned in **any** order.

**Example 1:**

**Input:** nums1 = [1,2,3], nums2 = [2,4,6]

**Output:** [[1,3],[4,6]]

</aside>
```

```python
def find_missing_elements(nums1, nums2):
    # Convert both lists to sets to find the distinct elements
    set_nums1 = set(nums1)
    set_nums2 = set(nums2)

    # Find the distinct elements present in nums1 but not in nums2
    missing_in_nums2 = list(set_nums1 - set_nums2)

    # Find the distinct elements present in nums2 but not in nums1
    missing_in_nums1 = list(set_nums2 - set_nums1)

    # Return the lists of missing elements
    return [missing_in_nums1, missing_in_nums2]

# Test the function
nums1 = [1, 2, 3]
nums2 = [2, 4, 6]
print(find_missing_elements(nums1, nums2))  # Output: [[1, 3], [4, 6]]
```

```python
<aside>
ðŸ’¡ **Question 5**

Given two integer arrays arr1 and arr2, and the integer d, *return the distance value between the two arrays*.

The distance value is defined as the number of elements arr1[i] such that there is not any element arr2[j] where |arr1[i]-arr2[j]| <= d.

**Example 1:**

**Input:** arr1 = [4,5,8], arr2 = [10,9,1,8], d = 2

**Output:** 2

</aside>
```

```python
def distance_value(arr1, arr2, d):
    count = 0

    for num1 in arr1:
        has_close_element = False
        for num2 in arr2:
            if abs(num1 - num2) <= d:
                has_close_element = True
                break

        if not has_close_element:
            count += 1

    return count

# Test the function
arr1 = [4, 5, 8]
arr2 = [10, 9, 1, 8]
d = 2
print(distance_value(arr1, arr2, d))  # Output: 2
```

```python
<aside>
ðŸ’¡ **Question 6**

Given an integer array nums of length n where all the integers of nums are in the range [1, n] and each integer appears **once** or **twice**, return *an array of all the integers that appears **twice***.

You must write an algorithm that runs in O(n) time and uses only constant extra space.

**Example 1:**

**Input:** nums = [4,3,2,7,8,2,3,1]

</aside>
```

```python
def find_duplicates(nums):
    n = len(nums)
    duplicates = []

    i = 0
    while i < n:
        # The value at nums[i] should be placed at index (nums[i] - 1)
        correct_index = nums[i] - 1

        if nums[i] != nums[correct_index]:
            # Swap the elements to put nums[i] in its correct position
            nums[i], nums[correct_index] = nums[correct_index], nums[i]
        else:
            # If nums[i] is already in its correct position and not already marked as a duplicate, it is a duplicate.
            if i != correct_index:
                duplicates.append(nums[i])
            i += 1

    return duplicates

# Test the function
nums = [4, 3, 2, 7, 8, 2, 3, 1]
print(find_duplicates(nums))  # Output: [2, 3]
```

```python
<aside>
ðŸ’¡ **Question 7**

Suppose an array of length n sorted in ascending order is **rotated** between 1 and n times. For example, the array nums = [0,1,2,4,5,6,7] might become:

- [4,5,6,7,0,1,2] if it was rotated 4 times.
- [0,1,2,4,5,6,7] if it was rotated 7 times.

Notice that **rotating** an array [a[0], a[1], a[2], ..., a[n-1]] 1 time results in the array [a[n-1], a[0], a[1], a[2], ..., a[n-2]].

Given the sorted rotated array nums of **unique** elements, return *the minimum element of this array*.

You must write an algorithm that runs in O(log n) time.

**Example 1:**

**Input:** nums = [3,4,5,1,2]

**Output:** 1

**Explanation:**

The original array was [1,2,3,4,5] rotated 3 times.

</aside>
```

```python
def find_minimum_element(nums):
    left, right = 0, len(nums) - 1

    while left < right:
        mid = left + (right - left) // 2

        if nums[mid] > nums[right]:
            # The minimum element is in the right subarray.
            left = mid + 1
        else:
            # The minimum element is in the left subarray or is the mid element itself.
            right = mid

    return nums[left]

# Test the function
nums = [3, 4, 5, 1, 2]
print(find_minimum_element(nums))  # Output: 1
```

```python
<aside>
ðŸ’¡ **Question 8**

An integer array original is transformed into a **doubled** array changed by appending **twice the value** of every element in original, and then randomly **shuffling** the resulting array.

Given an array changed, return original *if* changed *is a **doubled** array. If* changed *is not a **doubled** array, return an empty array. The elements in* original *may be returned in **any** order*.

**Example 1:**

**Input:** changed = [1,3,4,2,6,8]

**Output:** [1,3,4]

**Explanation:** One possible original array could be [1,3,4]:

- Twice the value of 1 is 1 * 2 = 2.
- Twice the value of 3 is 3 * 2 = 6.
- Twice the value of 4 is 4 * 2 = 8.

Other original arrays could be [4,3,1] or [3,1,4].

</aside>
```

```python
def find_original(changed):
    original_map = {}

    # Count the occurrences of each element in the changed array and store their doubled values in the original_map.
    for num in changed:
        if num in original_map:
            original_map[num] += 1
        else:
            original_map[num] = 1

    # Process each element and its doubled value.
    for num in changed:
        doubled_value = num * 2

        # Check if both the element and its doubled value exist in the original_map.
        if num in original_map and doubled_value in original_map:
            original_map[num] -= 1
            original_map[doubled_value] -= 1

            # Remove elements with zero occurrences from the original_map.
            if original_map[num] == 0:
                del original_map[num]
            if original_map[doubled_value] == 0:
                del original_map[doubled_value]
        else:
            # If either the element or its doubled value is missing, changed is not a doubled array.
            return []

    # If the original_map is empty, changed is a valid doubled array. Return the keys as the original array.
    return list(original_map.keys())

# Test the function
changed = [1, 3, 4, 2, 6, 8]
print(find_original(changed))  # Output: [1, 3, 4]
print([1, 3, 4])
```
