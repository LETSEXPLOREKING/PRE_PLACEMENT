```python
<aside>
ðŸ’¡ **Question 1**

Given two strings s1 and s2, return *the lowest **ASCII** sum of deleted characters to make two strings equal*.

**Example 1:**

**Input:** s1 = "sea", s2 = "eat"

**Output:** 231

**Explanation:** Deleting "s" from "sea" adds the ASCII value of "s" (115) to the sum.

Deleting "t" from "eat" adds 116 to the sum.

At the end, both strings are equal, and 115 + 116 = 231 is the minimum sum possible to achieve this.

</aside>
```

```python
def minimum_ascii_sum_deleted_chars(s1, s2):
    # Function to calculate the ASCII value of a string
    def ascii_sum(s):
        return sum(ord(char) for char in s)

    # Lengths of the input strings
    m, n = len(s1), len(s2)

    # Create a 2D DP array to store the minimum ASCII sum of deleted characters
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Initialize the DP array
    for i in range(m - 1, -1, -1):
        dp[i][n] = dp[i + 1][n] + ord(s1[i])

    for j in range(n - 1, -1, -1):
        dp[m][j] = dp[m][j + 1] + ord(s2[j])

    # Fill the DP array using bottom-up dynamic programming approach
    for i in range(m - 1, -1, -1):
        for j in range(n - 1, -1, -1):
            if s1[i] == s2[j]:
                dp[i][j] = dp[i + 1][j + 1]
            else:
                dp[i][j] = min(ord(s1[i]) + dp[i + 1][j], ord(s2[j]) + dp[i][j + 1])

    return dp[0][0]

# Example usage:
s1 = "sea"
s2 = "eat"
print(minimum_ascii_sum_deleted_chars(s1, s2))  # Output: 231
```

```python
<aside>
ðŸ’¡ **Question 2**

Given a string s containing only three types of characters: '(', ')' and '*', return true *if* s *is **valid***.

The following rules define a **valid** string:

- Any left parenthesis '(' must have a corresponding right parenthesis ')'.
- Any right parenthesis ')' must have a corresponding left parenthesis '('.
- Left parenthesis '(' must go before the corresponding right parenthesis ')'.
- '*' could be treated as a single right parenthesis ')' or a single left parenthesis '(' or an empty string "".

**Example 1:**

**Input:** s = "()"

**Output:**

true

</aside>
```

```python
def is_valid_string(s):
    stack = []
    star_stack = []

    for i, char in enumerate(s):
        if char == '(':
            stack.append(i)
        elif char == '*':
            star_stack.append(i)
        else:
            if stack:
                stack.pop()
            elif star_stack:
                star_stack.pop()
            else:
                return False

    while stack and star_stack:
        if stack[-1] > star_stack[-1]:
            return False
        stack.pop()
        star_stack.pop()

    return not stack

# Example usage:
s1 = "()"
print(is_valid_string(s1))  # Output: True
```

```python
<aside>
ðŸ’¡ **Question 3**

Given two strings word1 and word2, return *the minimum number of **steps** required to make* word1 *and* word2 *the same*.

In one **step**, you can delete exactly one character in either string.

**Example 1:**

**Input:** word1 = "sea", word2 = "eat"

**Output:** 2

**Explanation:** You need one step to make "sea" to "ea" and another step to make "eat" to "ea".

</aside>
```

```python
def min_steps_to_same(word1, word2):
    m, n = len(word1), len(word2)

    # Create a 2D DP array to store the minimum steps to make substrings equal
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Initialize the DP array
    for i in range(m - 1, -1, -1):
        dp[i][n] = dp[i + 1][n] + 1

    for j in range(n - 1, -1, -1):
        dp[m][j] = dp[m][j + 1] + 1

    # Fill the DP array using bottom-up dynamic programming approach
    for i in range(m - 1, -1, -1):
        for j in range(n - 1, -1, -1):
            if word1[i] == word2[j]:
                dp[i][j] = dp[i + 1][j + 1]
            else:
                dp[i][j] = min(dp[i + 1][j], dp[i][j + 1]) + 1

    return dp[0][0]

# Example usage:
word1 = "sea"
word2 = "eat"
print(min_steps_to_same(word1, word2))  # Output: 2
```

```python
<aside>
ðŸ’¡ **Question 4**

You need to construct a binary tree from a string consisting of parenthesis and integers.

The whole input represents a binary tree. It contains an integer followed by zero, one or two pairs of parenthesis. The integer represents the root's value and a pair of parenthesis contains a child binary tree with the same structure.
You always start to construct the **left** child node of the parent first if it exists.

</aside>**Input:** s = "4(2(3)(1))(6(5))"

**Output:** [4,2,6,3,1,5]
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def str2tree(s):
    # Helper function to parse the integer value from the string
    def parse_value(start):
        end = start
        while end < len(s) and (s[end] == '-' or s[end].isdigit()):
            end += 1
        return int(s[start:end]), end

    # Recursive function to construct the binary tree from the string
    def build_tree(start):
        if start >= len(s):
            return None, start

        if s[start] == '(':
            start += 1

        if s[start] == ')':
            return None, start + 1

        value, start = parse_value(start)
        root = TreeNode(value)

        root.left, start = build_tree(start)

        if start < len(s) and s[start] == '(':
            root.right, start = build_tree(start)

        if start < len(s) and s[start] == ')':
            start += 1

        return root, start

    # Call the recursive function with the start index as 0
    root, _ = build_tree(0)
    return root

# Helper function to convert the binary tree to a list in pre-order traversal
def tree_to_list(root):
    if not root:
        return []
    return [root.val] + tree_to_list(root.left) + tree_to_list(root.right)

# Example usage:
s = "4(2(3)(1))(6(5))"
tree = str2tree(s)
result_list = tree_to_list(tree)
print(result_list)  # Output: [4, 2, 3, 1, 6, 5]
```

```python
<aside>
ðŸ’¡ **Question 5**

Given an array of characters chars, compress it using the following algorithm:

Begin with an empty string s. For each group of **consecutive repeating characters** in chars:

- If the group's length is 1, append the character to s.
- Otherwise, append the character followed by the group's length.

The compressed string s **should not be returned separately**, but instead, be stored **in the input character array chars**. Note that group lengths that are 10 or longer will be split into multiple characters in chars.

After you are done **modifying the input array,** return *the new length of the array*.

You must write an algorithm that uses only constant extra space.

**Example 1:**

**Input:** chars = ["a","a","b","b","c","c","c"]

**Output:** Return 6, and the first 6 characters of the input array should be: ["a","2","b","2","c","3"]

**Explanation:**

The groups are "aa", "bb", and "ccc". This compresses to "a2b2c3".

</aside>
```

```python
def compress(chars):
    # Function to convert a count into a list of characters
    def convert_to_chars(count):
        if count == 1:
            return []
        return list(str(count))

    write_idx = 0
    read_idx = 0
    n = len(chars)

    while read_idx < n:
        char = chars[read_idx]
        count = 0

        while read_idx < n and chars[read_idx] == char:
            read_idx += 1
            count += 1

        chars[write_idx] = char
        write_idx += 1

        count_chars = convert_to_chars(count)
        for c in count_chars:
            chars[write_idx] = c
            write_idx += 1

    return write_idx

# Example usage:
chars = ["a", "a", "b", "b", "c", "c", "c"]
new_length = compress(chars)
print(new_length)  # Output: 6
print(chars[:new_length])  # Output: ["a", "2", "b", "2", "c", "3"]
```

```python
<aside>
ðŸ’¡ **Question 6**

Given two strings s and p, return *an array of all the start indices of* p*'s anagrams in* s. You may return the answer in **any order**.

An **Anagram** is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

**Example 1:**

**Input:** s = "cbaebabacd", p = "abc"

**Output:** [0,6]

**Explanation:**

The substring with start index = 0 is "cba", which is an anagram of "abc".

The substring with start index = 6 is "bac", which is an anagram of "abc".

</aside>
```

```python
def find_anagrams(s, p):
    result = []
    p_freq = {}
    for char in p:
        p_freq[char] = p_freq.get(char, 0) + 1

    window_freq = {}
    left, right = 0, 0
    matches = 0

    while right < len(s):
        char = s[right]
        if char in p_freq:
            window_freq[char] = window_freq.get(char, 0) + 1
            if window_freq[char] == p_freq[char]:
                matches += 1

        if right - left + 1 == len(p):
            if matches == len(p_freq):
                result.append(left)

            if s[left] in p_freq:
                window_freq[s[left]] -= 1
                if window_freq[s[left]] < p_freq[s[left]]:
                    matches -= 1

            left += 1

        right += 1

    return result

# Example usage:
s = "cbaebabacd"
p = "abc"
print(find_anagrams(s, p))  # Output: [0, 6]
```

```python
<aside>
ðŸ’¡ **Question 7**

Given an encoded string, return its decoded string.

The encoding rule is: k[encoded_string], where the encoded_string inside the square brackets is being repeated exactly k times. Note that k is guaranteed to be a positive integer.

You may assume that the input string is always valid; there are no extra white spaces, square brackets are well-formed, etc. Furthermore, you may assume that the original data does not contain any digits and that digits are only for those repeat numbers, k. For example, there will not be input like 3a or 2[4].

The test cases are generated so that the length of the output will never exceed 105.

**Example 1:**

**Input:** s = "3[a]2[bc]"

**Output:** "aaabcbc"

</aside>
```

```python
def decode_string(s):
    stack = []
    current_str = ""
    current_num = 0

    for char in s:
        if char.isdigit():
            current_num = current_num * 10 + int(char)
        elif char == '[':
            stack.append((current_str, current_num))
            current_str, current_num = "", 0
        elif char == ']':
            prev_str, prev_num = stack.pop()
            current_str = prev_str + prev_num * current_str
        else:
            current_str += char

    return current_str

# Example usage:
s = "3[a]2[bc]"
print(decode_string(s))  # Output: "aaabcbc"
```

```python
<aside>
ðŸ’¡ **Question 8**

Given two strings s and goal, return true *if you can swap two letters in* s *so the result is equal to* goal*, otherwise, return* false*.*

Swapping letters is defined as taking two indices i and j (0-indexed) such that i != j and swapping the characters at s[i] and s[j].

- For example, swapping at indices 0 and 2 in "abcd" results in "cbad".

**Example 1:**

**Input:** s = "ab", goal = "ba"

**Output:** true

**Explanation:** You can swap s[0] = 'a' and s[1] = 'b' to get "ba", which is equal to goal.

</aside>
```

```python
def can_swap_strings(s, goal):
    # Step 1: Find the differing indices
    diff_indices = [i for i in range(len(s)) if s[i] != goal[i]]

    # Step 2: Check if there are exactly two differing indices
    if len(diff_indices) == 2:
        i, j = diff_indices
        # Check if swapping characters at i and j in s results in goal
        if s[i] == goal[j] and s[j] == goal[i]:
            return True

    return False

# Example usage:
s = "ab"
goal = "ba"
print(can_swap_strings(s, goal))  # Output: True
```
